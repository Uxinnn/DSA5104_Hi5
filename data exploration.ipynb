```python
import json
```

```python
def start_requests():
    print("Start Trip Advisor spider...")
    with open("data/cmu/offering.txt", 'r') as f:
        for hotel_data in f:
            hotel_data = json.loads(hotel_data.strip())
            url = hotel_data["url"]
            yield url
#         url = next(self.urls)
#         logger.info(f"Crawling {url}")
#         yield scrapy.Request(url=url, callback=self.parse)

gen = start_requests()
gen
```
