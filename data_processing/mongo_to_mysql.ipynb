# Host

```python
host = pd.read_json('dsa5104.user.json')
host
```

```python
host_details = pd.json_normalize(host.host)
host_details = host_details.reindex(columns = ['host_since', 'host_address', 'host_responsetime','host_neighbourhood', 'host_verification','host_description'])
host_details
```

```python
hosts = host.reindex(columns = ['_id'])
hosts = hosts.rename(columns = {'_id':'uid'})
hosts = hosts.merge(host_details, left_index=True, right_index=True)
hosts = hosts.dropna(axis=0, how='all', subset=['host_since', 'host_address', 'host_responsetime','host_neighbourhood', 'host_verification','host_description'], ignore_index=True)
hosts
```

```python
des_list = []
for i in range(len(hosts)):
    a = hosts.host_description[i]
    if type(a)==type(str()):
        des_list.append(a.replace('\r\n', ''))
    else:
        des_list.append(a)
hosts['host_description']=des_list
hosts.to_csv('host.csv',sep='@', index=False, lineterminator='\r\n' )
hosts
```

# Review

```python
reviews = pd.read_json('dsa5104.reviews.json')
reviews = reviews.rename(columns = {'_id': 'rid', 'reviewer_id': 'uid', 'listing_id':'aid','date':'review_date'})
reviews = reviews.reindex(columns = ['rid','uid', 'aid','title','text', 'review_date', 'overall_score','cleanliness_score','service_score','value_score','location_score','sleepquality_score', 'checkin_score','communication_score'])
reviews
```

```python
rid_list, uid_list, aid_list = [], [], []
for i in range(len(reviews)):
    r = reviews.rid[i]
    u = reviews.uid[i]
    a = reviews.aid[i]
    if type(r)==type(dict()):
        rid_list.append(r['$numberLong'])
    else:
        rid_list.append(r)
    if type(u)==type(dict()):
        uid_list.append(u['$numberLong'])
    else:
        uid_list.append(u)
    if type(a)==type(dict()):
        aid_list.append(a['$numberLong'])
    else:
        aid_list.append(a)
reviews['rid']=rid_list
reviews['uid']=uid_list
reviews['aid']=aid_list

reviews = reviews.astype(str).apply(lambda x: x.str.encode('ascii', 'ignore').str.decode('ascii'))

reviews.to_csv('reviews.csv',sep='@', index=False, lineterminator='\r\n' )
reviews
```

# Transactions

```python
transactions = pd.read_json('dsa5104.transactions.json')
transactions = transactions.rename(columns={'_id':'tid','review_id':'rid', 'accomodation_id':'aid'})
transactions['date_start'] = pd.to_datetime(transactions['date_start'])
transactions['date_end'] = pd.to_datetime(transactions['date_end'])
transactions = transactions.reindex(columns = ['tid','rid','aid', 'price','date_start','date_end'])
transactions
```

```python
rid_list, uid_list, aid_list = [], [], []
for i in range(len(transactions)):
    r = transactions.rid[i]
    u = transactions.tid[i]
    a = transactions.aid[i]
    if type(r)==type(dict()):
        rid_list.append(r['$numberLong'])
    else:
        rid_list.append(r)
    if type(u)==type(dict()):
        uid_list.append(u['$oid'])
    else:
        uid_list.append(u)
    if type(a)==type(dict()):
        aid_list.append(a['$numberLong'])
    else:
        aid_list.append(a)
transactions['rid']=rid_list
transactions['tid']=uid_list
transactions['aid']=aid_list
transactions.to_csv('transactions.csv',sep='@', index=False, lineterminator='\r\n' )
transactions
```
